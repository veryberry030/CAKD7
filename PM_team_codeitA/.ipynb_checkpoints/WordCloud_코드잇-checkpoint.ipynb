## 워드클라우드 만들기

```python
text_data = df.Text.values
text_data
```

```python
data = ' '.join(text_data)
data
```

```python
import re

wf = re.sub('[^가-힣]',' ',data)
wf = re.sub('\s+',' ',wf)
wf
```

```python
#문자 형태 분석, tag:품시
from konlpy.tag import Okt

# Okt 함수를 이용해 형태소 분석
okt = Okt()
line =[]
line = okt.pos(wf)

noun_list =[]
# 명사 또는 형용사인 단어만 noun_list에 넣어주기
for word, tag in line:
    if(tag=='Noun') & (len(word)>1):
        noun_list.append(word)
# 명사 또는 형용사인 단어 및 2글자 이상인 단어 선택 시
# n_adj = [word for word, tag in line if tag in ['Noun','Adjective'] and len(word) > 1]

#제외할 단어 추가
stop_words = "디스 강의 큐알 드레스 진짜 존나 라프텔 아마 그거 요즘 자꾸 여기 자기 유치 흐앙 케틀벨 헬스 아옮 서여 스시 뜰때 유미 혹시 수강 친추 오늘 사람 사이트 광고 지금 그냥 거기 다시 아 코드잇 정말 분씩 하고 있어요 년 건 한 쿠키뉴스 뉴스 흑흑 헤헷 헤헥 코드 드레스코드 친구코드 디스코드 이벤트코드 큐알코드 중복코드"
stop_words = set(stop_words.split(' '))
        
noun_list = [word for word in noun_list if not word in stop_words]
print(noun_list)
```

```python
from collections import Counter
one_word = []
for word, tag in line:
    if (tag == 'Noun') & (len(word)==4):
        one_word.append(word)
one_dict = dict(Counter(one_word).most_common())
keys = sorted(one_dict.items(), key=lambda x:x[1], reverse=True)
keys[:10]
```

```python
df_keys = pd.DataFrame(keys[:10])
df_keys = df_keys.rename(columns={0:'keyword',1:'freq'})
df_keys
```

```python
from collections import Counter
from matplotlib import rc
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

rc('font',family='Malgun Gothic') #한글 폰트 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 부호 출력 설정
count = Counter(noun_list)
data = dict(count.most_common())
series = pd.Series(data)
df = pd.DataFrame(series,columns=['cnt'])
df1 = df.sort_values(by='cnt',ascending=False)[:50]
df1.reset_index(inplace=True)
df1
 
```

```python
from wordcloud import WordCloud

plt.figure(figsize=(16,12))
wordcloud = WordCloud(font_path='./BMHANNAPro.ttf',background_color='black',colormap='Pastel1',
                      width=800, height=600, max_words=60, prefer_horizontal=1).generate_from_frequencies(data)
plt.imshow(wordcloud)
plt.axis('off')
```

```python
wordcloud.to_file('codeit_tweeter1.png')
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np

def grey_color_func(word, font_size, position,orientation,random_state=None, **kwargs):
    return("hsl({:d},{:d}%, {:d}%)".format(np.random.randint(212,313),np.random.randint(26,32),np.random.randint(45,80)))
    #return("hsl(230,100%%, %d%%)" % np.random.randint(49,51))

plt.figure(figsize=(16,12))
wordcloud = WordCloud(font_path='./BMHANNAPro.ttf',background_color='black',prefer_horizontal=1,
                      width=600,max_words=60, height=400).generate_from_frequencies(data)

#change the color setting
wordcloud.recolor(color_func = grey_color_func)

plt.imshow(wordcloud)
plt.axis('off')
```

```python
import re

def text_preprocessing(x):
    data = x
    wf = re.sub('[^가-힣]',' ',data)
    wf = re.sub('\s+',' ',wf)
    return wf

df.Text = df.Text.apply(lambda x:text_preprocessing(x))
df['Datetime'] = df['Datetime'].apply(lambda x:x[:11])
df = df[['Datetime','Text']]
df
```
