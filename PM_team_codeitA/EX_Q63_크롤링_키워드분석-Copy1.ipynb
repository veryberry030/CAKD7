# 형태소 분석

```python
import konlpy
from konlpy.tag import Okt
from collections import Counter
from matplotlib import rc
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
okt = Okt()
morph = okt.pos(text)
```

```python
noun_list = []
for word, tag in morph:
    if (tag == 'Noun') and (len(word) > 1):
        noun_list.append(word)
print(noun_list)
```

```python
count = Counter(noun_list)
data = dict(count.most_common())
data
```

```python
import pickle
```

```python
with open('data.pkl', 'wb') as fw:
    pickle.dump(data,fw)
```

# 불용어 처리 stop_words

```python
data
```

```python
stopwords = "익스 아이 머니투데이 바로가기 이데일리 경우 이하 각종 가운데 정원 통해 훈련 디지털 과정 투자 분야 창업 억원 제공 선정 중장 만원 개사 최종 현재 최근 진행 경제 위해 발급"
```

```python
stopwords = stopwords.split(" ")
```

```python
data1={}
for word in data:
    if not word in stopwords:
        data1[word] = data[word]
```
