```python
req = requests.get(url).content
xmlobject = xmltodict.parse(req)
xmlobject['LawSearch']['law']
```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(xmlobject['LawSearch']['law'])
df.head()
```

```python
target = df['법령명한글'].values
title_list = target.tolist()#배열을 리스트로 변환
text = ' '.join(title_list)#타이틀 리스트를 공백문자 하나로 이어줌
text
```

```python
import re
text = re.sub('[^가-힣]+',' ',text)
print(len(text))
text[:100]
```

```python
from konlpy.tag import Okt
#문자 형태 분석, tag:품시

okt = Okt()
morph = okt.pos(text) #형태소 변환
noun_list = []
for word, tag in morph:
    if(tag=='Noun') & (len(word)>1):
        noun_list.append(word)
print(morph)
print(noun_list)
```

```python
from collections import Counter
one_word = []
for word, tag in morph:
    if (tag == 'Noun') & (len(word)==1):
        one_word.append(word)
one_dict = dict(Counter(one_word).most_common())
keys = sorted(one_dict.items(), key=lambda x:x[1], reverse=True)
keys[:10]
```

```python
from collections import Counter
from matplotlib import rc
import matplotlib.pyplot as plt
import seaborn as sns

rc('font',family='Malgun Gothic') #한글 폰트 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 부호 출력 설정
count = Counter(noun_list)
data = dict(count.most_common())
serise = pd.Series(data)
df = pd.DataFrame(serise,columns=['cnt'])
df10 = df.sort_values(by='cnt',ascending=False)[:10]
df10.reset_index(inplace=True)
df10
```

```python
sns.barplot(x='index',y='cnt',data=df10, orient='v')
```

# 과제 : 관심있는 단어로 워드클라우드 만들어보기

```python
# url 읽어오기
url = 'https://search.naver.com/search.naver?where=news&query=AI&sm=tab_tmr&nso=so:r,p:all,a:all&sort=0'
URL='http://naver.com'
response = requests.get(URL)
html_data = response.text
html_data
```

```python
# 한글만 추출
import re

wf = re.sub('[^가-힣]',' ',html_data)
wf
```

```python
#공백 제거
wf = re.sub('\s+',' ',wf)
wf
```

```python
#문자 형태 분석, tag:품시
from konlpy.tag import Okt

okt = Okt()
morph = okt.pos(wf) #형태소 변환
noun_list = []
for word, tag in morph:
    if(tag=='Noun') & (len(word)>1):
        noun_list.append(word)
print(morph)
print(noun_list)
```

```python
from collections import Counter
one_word = []
for word, tag in morph:
    if (tag == 'Noun') & (len(word)==1):
        one_word.append(word)
one_dict = dict(Counter(one_word).most_common())
keys = sorted(one_dict.items(), key=lambda x:x[1], reverse=True)
keys[:10]
```

```python
from collections import Counter
from matplotlib import rc
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

rc('font',family='Malgun Gothic') #한글 폰트 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 부호 출력 설정
count = Counter(noun_list)
data = dict(count.most_common())
serise = pd.Series(data)
df = pd.DataFrame(serise,columns=['cnt'])
df10 = df.sort_values(by='cnt',ascending=False)[:10]
df10.reset_index(inplace=True)
df10
```

```python
sns.barplot(x='index',y='cnt',data=df10, orient='v')
```

```python
from wordcloud import WordCloud

wordcloud = WordCloud(font_path='../Visualization/dataset/malgun.ttf',background_color='white',
                     colormap='flare', width=1500, height=1000).generate_from_frequencies(data)
plt.imshow(wordcloud)
plt.axis('off')
```
