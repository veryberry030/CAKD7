### 도전과제
mtcars 데이터셋의 qsec 컬럼을 최소 최대 척도로 변환한 후 0.5보다 큰 값을 가지는 레코드 수를 구하시오.

```python
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv('dataset/mtcars.csv',index_col=0)

scaler = MinMaxScaler()
scaled = scaler.fit_transform(df)

X = df['qsec']
X_MinMax = (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))

overNum = len(X_MinMax[X_MinMax>0.5])

X_MinMax.values
```

```python
import seaborn as sns

df = sns.load_dataset('titanic')
df.head()
```

```python
df.age.value_counts()
```

```python
# 범주화 하기
def get_category(x):
    cat = ''
    if x < 20:
        cat = 'child'
    elif x < 30:
        cat = 'Young Adult'
    elif x < 50:
        cat = 'Adult'
    else:
        cat = 'Elderly'
    
    return cat
            
df['age_class'] = df.age.apply(lambda x:get_category(x))
df.head()
```

```python
#소문자를 대문자로 변환

import math

def upper(x):
    return x.upper()

df['sex'] = df.sex.apply(lambda x:upper(x))
df[['sex']].head()
```
