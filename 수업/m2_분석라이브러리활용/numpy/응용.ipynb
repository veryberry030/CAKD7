[numpy random 함수]

- np.random.seed : seed를 통한 난수 생성
- np.random.randint : 정수 난수 1개 생성
- np.random.rand : 0부터 1사이의 균일분포에서 난수 매트릭스 배열 생성
- np.random.randn : 가우시안 표준 정규분포에서 난수 매트릭스 배열 생성
- np.random.shuffle : 기존의 데이터의 순서 바꾸기
- np.random.choice : 기존 데이터에서 sampling


```python
# python random
import random

print(random.random(),'\n') # 0.0 <= x < 1.0 사이 랜덤한 숫자

print(random.randint(1,10),'\n') # 1 <= x <=10 사이 랜덤한 정수

print(random.uniform(10,20),'\n') # min,max 사이 float

print(random.randrange(10),'\n') # 지정범위 int 리턴

print(random.choice([1,2,3,4,5]),'\n') # 리스트 내부에 있는 요소를 랜덤하게 선택

li = [1,2,3,4,5]
print(random.sample(li,3),'\n') # 리스트 요소를 중복이 안되게 리턴

random.shuffle(li) #리스트 요소를 섞기
print(li)
```

### Q1. 고객 속성(id, 성별, 연령, 거주지역) 데이터와 거래 데이터를 통합한 데이터 프레임 df를 생성하세요.


```python
# 고객별 속성 데이터

import pandas as pd
import numpy as np

np.random.seed(1)

id = np.arange(1,1001)
id1 = pd.Series(id)

cusno = np.random.randint(1001,size=1000)
cn1 = pd.Series(cusno)

gender = np.random.randint(2,size=1000)
g1 = pd.Series(gender)

age = np.random.randint(81,size=1000)
a1 = pd.Series(age)

region = np.random.randint(11,size=1000)
r1 = pd.Series(region)


df_cus = pd.concat([id1,cn1,g1,a1,r1],axis=1)
df_cus.rename(columns={0:'id',1:'cusno',2:'gender',3:'age',4:'region'},inplace=True)

df_cus.index = df_cus.index+1

df_cus
```

```python
# 거래 데이터
# product_ : 고가제품(10), 중가제품(20), 저가제품(30)로 구분제품코드 10,20,30
# sales : 20년 대비 21년구매금액이 증가면 1, 감소면 0 


    
import pandas as pd
import numpy as np

np.random.seed(1)

product_20 = np.random.choice([10,20,30],size=1000)
pd0 = pd.Series(product_20)

product_21 = np.random.choice([10,20,30],size=1000)
pd1 = pd.Series(product_21)

price_20 = np.random.randint(1000,50000,size=1000)
pr0 = pd.Series(price_20)

price_21 = np.random.randint(1000,50000,size=1000)
pr1 = pd.Series(price_21)

quantity_20 = np.random.randint(1,101,size=1000)
q0 = pd.Series(quantity_20)

quantity_21 = np.random.randint(1,101,size=1000)
q1 = pd.Series(quantity_21)

amount_20 = price_20 * quantity_20
am0 = pd.Series(amount_20)

amount_21 = price_21 * quantity_21
am1 = pd.Series(amount_21)

sales = np.random.randint(2,size=1000)
s1 = pd.Series(sales)


df_tra = pd.concat([pd0,pd1,pr0,pr1,q0,q1,am0,am1,s1],axis=1)
df_tra.rename(columns={0:'product_20',1:'product_21',2:'price_20',3:'price_21',4:'quantity_20',5:'quantity_21',
                       6:'amount_20',7:'amount_21',8:'sales'},inplace=True)

df_tra.index = df_tra.index+1

df_tra
```

```python
df = pd.concat([df_cus,df_tra],axis=1)
df
```

```python
df['sales'] = df.sales.apply(lambda x:1 if x > 0 else 0)
df
```

```python
df['sales'] = np.where(df.amount_20<df.amount_21,1,0)
```

```python
com = df['amount_20']<df['amount_21']

def tf(x):
    if x==True:
        return 1
    else:
        return 0
    
df['sales'] = com    
df['sales'] = df['sales'].apply(lambda x:tf(x))

df
```

```python
# 나이대별 클래스 분류
def age_class(x):
    if x<10:
        x = 'Child'
        
    elif 10 <= x < 20:
        x = 'Young Child'
        
    elif 20 <= x < 30:
        x = 'Young Adult'
        
    elif 30 <= x < 50:
        x = 'Adult'
        
    else :
        x = 'Senior'
    
    return x
        
df['age_class'] = df.age.apply(lambda x:age_class(x))
df
```
