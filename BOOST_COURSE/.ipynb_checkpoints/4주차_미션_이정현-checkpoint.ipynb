# üìå Q1. 
```
ÏãúÏ†ê Ïª¨ÎüºÏóêÏÑú Ïó∞ÎèÑÏôÄ Î∂ÑÍ∏∞Ïóê ÎåÄÌïú ÌååÏÉùÎ≥ÄÏàòÎ•º ÏÉùÏÑ±ÌïòÍ≥† Í∏∞Ïà†ÌÜµÍ≥ÑÎ•º Íµ¨Ìï¥Ï£ºÏÑ∏Ïöî.

- Îç∞Ïù¥ÌÑ∞ Ïª¨ÎüºÏùÑ Îã®ÏúÑÏù∏ "Î∞±Îßå"ÏúºÎ°ú Î≥ÄÍ≤ΩÌï¥ Ï£ºÏÑ∏Ïöî.
describe() Î°ú Í∏∞Ïà†ÌÜµÍ≥ÑÎ•º Íµ¨ÌñàÏùÑ Îïå Îã§ÏùåÍ≥º Í∞ôÏùÄ Í≤∞Í≥ºÍ∞Ä ÎÇòÏò§ÎèÑÎ°ù Ìï©ÎãàÎã§. 
ÏïÑÎûò Í≤∞Í≥ºÎäî ÏòàÏãúÎ°ú ÏÜåÏà´Ï†ê ÏûêÎ¶¨ÏàòÍ∞Ä Îã§Î•¥Í≤å ÌëúÏãúÎêòÎçîÎùºÎèÑ Í∞íÏù¥ Í∞ôÎã§Î©¥ Í¥úÏ∞ÆÏäµÎãàÎã§.

- Ï†ÑÏ≤òÎ¶¨Ìïú Î∞±Îßå, Ïó∞ÎèÑ, Î∂ÑÍ∏∞ Ïª¨ÎüºÏù¥ ÏàòÏπòÌòï Îç∞Ïù¥ÌÑ∞ Ïùº Îïå ÏïÑÎûòÏôÄ Í∞ôÏùÄ Í≤∞Í≥ºÍ∞Ä ÎÇòÏò§Îãà Îç∞Ïù¥ÌÑ∞Í∞Ä ÏàòÏπò ÌòïÌÉúÎ°ú ÎêòÏñ¥ ÏûàÎäîÏßÄÎèÑ ÌôïÏù∏Ìï¥ Ï£ºÏÑ∏Ïöî!

- Îç∞Ïù¥ÌÑ∞Îäî ÏúÑÏóêÏÑú Ï†úÍ≥µÎêú Îç∞Ïù¥ÌÑ∞Î•º ÏÇ¨Ïö©Ìï©ÎãàÎã§. Ìï¥Îãπ Í∏∞Ïà†ÌÜµÍ≥ÑÎ°ú Ïïå Ïàò ÏûàÎäî Ï†ïÎ≥¥Í∞Ä Î¨¥ÏóáÏù∏ÏßÄÎèÑ Ìï®Íªò ÏûëÏÑ±Ìï¥ Ï£ºÏÑ∏Ïöî!
```

````python
# Îç∞Ïù¥ÌÑ∞ Ïª¨ÎüºÏùÑ Îã®ÏúÑÏù∏ "Î∞±Îßå"ÏúºÎ°ú Î≥ÄÍ≤Ω
df = df.rename(columns={'Îç∞Ïù¥ÌÑ∞':'Î∞±Îßå'})
df.head()
````

````python
# Ïó∞ÎèÑ Ïª¨Îüº ÏÉùÏÑ±
df['Ïó∞ÎèÑ']=df['ÏãúÏ†ê'].map(lambda x:int(x[0:4]))
df.head()
````

````python
# Î∂ÑÍ∏∞ Ïª¨Îüº ÏÉùÏÑ±
df['Î∂ÑÍ∏∞']=df['ÏãúÏ†ê'].map(lambda x:int(x[5]))
df.head()
````

````python
# ÏàòÏπòÌòï ÏûêÎ£å Í∏∞Ïà† ÌÜµÍ≥Ñ
df.describe()
````

# üìå Q2. 

```
pivot_tableÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ Íµ≠Í∞Ä(ÎåÄÎ•ô)Î≥Ñ Ïó∞ÎèÑÎ≥Ñ ÌåêÎß§Ïï°Ïùò Ìï©Í≥ÑÎ•º Î∂ÑÏÑùÌï¥ Ï£ºÏÑ∏Ïöî.

Íµ≠Í∞Ä(ÎåÄÎ•ô)Î≥Ñ Ïó∞ÎèÑÎ≥Ñ ÌåêÎß§Ïï°Ïùò Ìï©Í≥ÑÎ•º Íµ¨Ìï¥ÏÑú Ïñ¥Îäê ÏßÄÏó≠Ïóê ÌåêÎß§Ïï°Ïù¥ ÎßéÏùÄÏßÄÎ•º ÏïåÏïÑÎ≥¥Í≤†ÏäµÎãàÎã§.

pivot_table ÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ Îã§ÏùåÏùò Í≤∞Í≥ºÎ•º Íµ¨ÌïòÍ≥† ÏãúÍ∞ÅÌôî Ìï¥Ï£ºÏÑ∏Ïöî!
Í∑∏ÎûòÌîÑÏùò ÏÉâÏÉÅ, Ïä§ÌÉÄÏùºÏùÄ ÏûêÏú†Î°≠Í≤å ÏõêÌïòÎäî Ïä§ÌÉÄÏùºÎ°ú ÏßÄÏ†ïÌï¥ÎèÑ Îê©ÎãàÎã§.
```

````python
# pivot_tableÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ Íµ≠Í∞Ä(ÎåÄÎ•ô)Î≥Ñ Ïó∞ÎèÑÎ≥Ñ ÌåêÎß§Ïï°Ïùò Ìï©Í≥Ñ Íµ¨ÌïòÍ∏∞
df_country = df.pivot_table(index='Íµ≠Í∞Ä(ÎåÄÎ•ô)Î≥Ñ',columns='Ïó∞ÎèÑ',values='Î∞±Îßå')
df_country
````

````python
# ÎùºÏù∏ÌîåÎ°Ø Í∑∏Î¶¨Í∏∞
df_country.plot.line()
````

# üìå Q3. 
```
groupbyÎ•º ÏÇ¨Ïö©ÌïòÏó¨ 2020ÎÖÑÏùò Ïò®ÎùºÏù∏ Ìï¥Ïô∏ÌåêÎß§ ÏÉÅÌíàÍµ∞ÏùÑ ÏßÄÏó≠Î≥Ñ Ìï©Í≥ÑÎ•º Íµ¨Ìï¥ Î∂ÑÏÑùÌï¥ Ï£ºÏÑ∏Ïöî.

2020ÎÖÑÏùÄ ÏΩîÎ°úÎÇòÎ°ú Ïù∏Ìï¥ Ï†ÑÍµ≠ÎØºÏù¥ Ïñ¥Î†§ÏõÄÏùÑ Í≤™ÏóàÎçò ÏãúÍ∏∞ÏûÖÎãàÎã§. Îòê, ÏïÑÎßàÏ°¥ Îì±Ïùò Ìï¥Ïô∏ Ïò®ÎùºÏù∏ ÌåêÎß§Í∞Ä ÎäòÏñ¥ÎÇú ÏãúÍ∏∞Ïù¥Í∏∞ÎèÑ Ìï©ÎãàÎã§. Ïù¥ Îïå, Ìï¥Ïô∏ Ïñ¥ÎäêÏßÄÏó≠Ïóê Ïò®ÎùºÏù∏ ÌåêÎß§Í∞Ä ÏñºÎßàÎÇò ÎêòÏóàÎäîÏßÄÎ•º groupby() ÏôÄ unstack()ÏùÑ ÌôúÏö©Ìï¥ÏÑú Ìï©Í≥ÑÎ•º Íµ¨ÌïòÍ≥† Î∂ÑÏÑùÌï¥ Ï£ºÏÑ∏Ïöî! Îòê Î∂ÑÏÑùÌïú Í≤∞Í≥ºÎ•º Î≥¥Í≥† Ïñ¥Îñ§ Ïù∏ÏÇ¨Ïù¥Ìä∏Î•º ÏñªÏóàÎäîÏßÄÎèÑ Ìï®Íªò ÏûëÏÑ±Ìï¥ Ï£ºÏÑ∏Ïöî!
Í∑∏ÎûòÌîÑÏùò ÏÉâÏÉÅ, Ïä§ÌÉÄÏùºÏùÄ ÏûêÏú†Î°≠Í≤å ÏõêÌïòÎäî Ïä§ÌÉÄÏùºÎ°ú ÏßÄÏ†ïÌï¥ÎèÑ Îê©ÎãàÎã§.
```

````python
df.columns
````

````python
# 2020ÎÖÑ Îç∞Ïù¥ÌÑ∞ Ï∂îÏ∂ú
df_2020 = df[df['Ïó∞ÎèÑ']==2020].copy()
df_2020
````

````python
# groupby() ÏôÄ unstack()ÏùÑ ÌôúÏö©Ìï¥ÏÑú Ìï©Í≥ÑÎ•º Íµ¨ÌïòÍ∏∞
df_q3 = df_2020.groupby(['ÏÉÅÌíàÍµ∞Î≥Ñ','Íµ≠Í∞Ä(ÎåÄÎ•ô)Î≥Ñ'])['Î∞±Îßå'].sum().unstack()
df_q3
````

````python
# ÌûàÌä∏ÎßµÏúºÎ°ú ÎßéÏù¥ Íµ≠Í∞ÄÎ≥Ñ ÎßéÏù¥ ÌåîÎ¶∞ ÏÉÅÌíàÍµ∞ ÏãúÍ∞ÅÌôî
plt.figure(figsize=(15,8))
sns.heatmap(data=df_q3,cmap='Blues',annot=True,fmt='.0f')
````

# üìå Q4. 
```
Ï£ºÏöî ÌåêÎß§ Íµ≠Í∞ÄÏôÄ ÏÉÅÌíàÍµ∞Ïóê ÎåÄÌï¥ 2021ÎÖÑÍπåÏßÄÏùò Ïó∞ÎèÑÎ≥Ñ Ïò®ÎùºÏù∏ ÏßÅÏ†ë ÌåêÎß§Ïï° Ìï©Í≥ÑÎ•º ÏãúÍ∞ÅÌôî Ìï¥Ï£ºÏÑ∏Ïöî.

Ïò®ÎùºÏù∏ ÏßÅÏ†ë ÌåêÎß§Ïï°Ïù¥ ÎÜíÏùÄ ÏßÄÏó≠ÏùÄ "ÎØ∏Íµ≠", "Ï§ëÍµ≠", "ÏùºÎ≥∏" Ïù¥Î©∞, Ï£ºÏöî ÌåêÎß§ ÏÉÅÌíàÍµ∞ÏùÄ "ÏùòÎ•ò Î∞è Ìå®ÏÖò Í¥ÄÎ†®ÏÉÅÌíà", "ÌôîÏû•Ìíà", "ÏùåÎ∞ò¬∑ÎπÑÎîîÏò§¬∑ÏïÖÍ∏∞" ÏûÖÎãàÎã§. 2021ÎÖÑÍπåÏßÄÏùò Í∞Å ÏßÄÏó≠Î≥Ñ Ï†úÌíàÎ≥Ñ ÌåêÎß§Ïï°ÏùÑ ÏãúÍ∞ÅÌôî Ìï¥Ï£ºÏÑ∏Ïöî!
ÏãúÍ∞ÅÌôî Í≤∞Í≥ºÏùò ÏÉâÏÉÅÏù¥ÎÇò Ïä§ÌÉÄÏùºÏù¥ÎÇò Ïä§ÌÉÄÏùºÏùÄ Îã¨ÎùºÎèÑ Í¥úÏ∞ÆÏäµÎãàÎã§.
```
 

````python
# Ï£ºÏöî ÌåêÎß§ Íµ≠Í∞ÄÏôÄ ÏÉÅÌíàÍµ∞ ÌôïÏù∏
df.groupby(['Íµ≠Í∞Ä(ÎåÄÎ•ô)Î≥Ñ','ÏÉÅÌíàÍµ∞Î≥Ñ']).sum().sort_values(by='Î∞±Îßå',ascending=False).head(10)
````

````python
# ÎØ∏Íµ≠, Ï§ëÍµ≠, ÏùºÎ≥∏Îßå Ï∂îÏ∂ú
df_q4 = df[(df['Íµ≠Í∞Ä(ÎåÄÎ•ô)Î≥Ñ']=='ÎØ∏Íµ≠')|(df['Íµ≠Í∞Ä(ÎåÄÎ•ô)Î≥Ñ']=='Ï§ëÍµ≠')|(df['Íµ≠Í∞Ä(ÎåÄÎ•ô)Î≥Ñ']=='ÏùºÎ≥∏')].copy()
df_q4
````

````python
df_q44 = df_q4[(df_q4['ÏÉÅÌíàÍµ∞Î≥Ñ']=='ÏùòÎ•ò Î∞è Ìå®ÏÖò Í¥ÄÎ†®ÏÉÅÌíà')|(df_q4['ÏÉÅÌíàÍµ∞Î≥Ñ']=='ÌôîÏû•Ìíà')|(df_q4['ÏÉÅÌíàÍµ∞Î≥Ñ']=='ÏùåÎ∞ò¬∑ÎπÑÎîîÏò§¬∑ÏïÖÍ∏∞')].copy()
df_q44
````

````python
# relplot ÏãúÍ∞ÅÌôî
sns.relplot(data=df_q44,x='Ïó∞ÎèÑ',y='Î∞±Îßå',hue='ÏÉÅÌíàÍµ∞Î≥Ñ',col='Íµ≠Í∞Ä(ÎåÄÎ•ô)Î≥Ñ',col_wrap=3,kind='line',errorbar=None)
````
