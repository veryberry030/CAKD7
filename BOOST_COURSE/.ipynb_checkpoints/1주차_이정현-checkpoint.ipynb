## 1.2 주피터 노트북 사용법
### 단축키 모음
  - enter : 편집상태로 변환
  - Shift + Enter : 셀이 실행이 되고 커서가 다음셀로 이동
  - Ctrl + Enter : 셀이 실행되고 커서가 그 셀에 유지
  - a : 위에 셀 한 개 추가
  - b : 아래 셀 한 개 추가
  - dd : 셀 삭제
  - m : 셀을 마크다운 셀로 변환
  - h : 다양한 주피터 노트북의 단축키를 확인
### 마크다운
  - 코드와 함께 문서화를 할 수 있음
  - 문서화를 할 수 있는 문법
### 주피터 노트북의 다양한 메뉴탭 활용
- View -> Toggle Line Numbers : 라인 넘버를 확인
- Kernel -> Restart & Clear Output : 커널과 결과를 초기화
- Nbextensions -> Table of Contents 의 기능을 활용하면 index 기능을 활용할 수 있음

## 2.1 데이터 분석을 위한 파이썬 속성 코스
- Zen of Python : import this 를 통해 파이썬의 철학을 알 수 있음
### 핵심 문법
- import
- boolean
  - True나 False는 0과 1로도 표현할 수 있으나 명시적으로 표현하기 위해 True와 False를 사용
- data type
- 반복문과 제어문 
    - for
        - enumerate를 사용하면 인덱스 번호와 원소를 같이 가져올 수 있음
          ex -> for i, val in enumerate(li)
    - if
- 문자열 관련
    - len : 문자열의 길이
    - split : 문자열 분리
    - list : 리스트 만들기
    - lower : 문자열 소문자로 만들기
    - upper : 문자열 대문자로 만들기
    - append : 문자열 붙이기
    - strip : 앞뒤 공백 제거
    
    - list[:2] : 슬라이싱으로 문자 가져오기
    - startswith : 특정 문자가 포함되는지 여부를 확인할 수 있음
    - in : 특정 문자열을 포함하고 있는지 여부를 확인할 수 있음
    - join : 리스트를 공백 문자열을 연결할 수 있음
